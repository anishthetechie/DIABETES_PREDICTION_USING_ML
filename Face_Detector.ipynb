```python
import numpy as np
import cv2
```

```python
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/anish_photo1.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_classifier.detectMultiScale(gray, 9,9)
if faces == ():
  print('Face not detected')
for (p,q,r,s) in faces:
  cv2.rectangle(img, (p,q), (p+r,q+s),(255,255,255),3)
  #(r,g,b), thickness
  cv2_imshow(img)
```
